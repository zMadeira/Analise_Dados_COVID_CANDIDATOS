```python
# Importando as bibliotecas.

%matplotlib inline

import pandas as pd
low_memory=False
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
# Importando o csv de dados do covid para a análise.
filename = r'C:\Users\gmadeira\OneDrive - Imagem Geosistemas e Comércio LTDA\Desktop\Guilherme\ep4\dados_covid.csv'
df_covid = pd.read_csv(filename, delimiter=';',error_bad_lines=False)


# Importando o csv de dados dos candidatos para a análise.
filename = 'dados_candidatos.csv'
df_candidatos = pd.read_csv(filename, delimiter=';', encoding='ISO-8859-1', error_bad_lines=False)
```

```python
# Gráfico de casos acumulados na região Sudeste do Brasil.
df_covid.groupby('municipio')['casosAcumulado'].max().sort_values().tail(5).plot(kind='barh', color='red',figsize =(10,5),label='Casos acumulados')
plt.title('Número de casos acumulados por Município na região Sudeste.')

plt.legend()
```

```python
# Gráfico de novos casos em São Paulo.
sp = df_covid.query('municipio == "São Paulo"')
sp.tail(18).plot(kind='line', 
        x='data', y='casosNovos', color='red',figsize =(10,5),label="Novos casos")

plt.title('Número de casos novos em São Paulo por data.')
plt.legend()
```

```python
# Gráfico da ocupação dos candidatos do Município de São Paulo.

df_candidatos.query('NM_UE == "SÃO PAULO"')['DS_OCUPACAO'].value_counts().head().plot(kind='barh',color='darkblue',figsize =(12,5))
plt.title('Ocupação dos candidatos do Município de São Paulo.')
```

```python
# Gráfico dos cargos dos candidatos do Município de São Paulo.

df_candidatos.query('NM_UE == "SÃO PAULO"')['DS_CARGO'].value_counts().head().plot(kind='barh',color='darkblue',figsize =(15,5),label="Candidatos")
plt.title('Cargo dos candidatos do Município de São Paulo.')

plt.legend()
```

```python
# Gráfico do gênero dos candidatos do Município de São Paulo.

df_candidatos.query('NM_UE == "SÃO PAULO"')['DS_GENERO'].value_counts().head().plot(kind='barh',figsize =(10,5),color='darkblue',label="Candidatos")
plt.title('Gênero dos candidatos do Município de São Paulo.')

plt.legend()
```

```python
# Gráfico da relação Idade de posse de cada partido.

sp = df_candidatos.query('NM_UE == "SÃO PAULO"')
sp.tail(18).plot(kind='bar', 
        x='SG_PARTIDO', y='NR_IDADE_DATA_POSSE', color='darkblue',figsize =(15,5),label="Idade")

plt.legend()

plt.title('Relação idade de posse de cada partido..')
```
